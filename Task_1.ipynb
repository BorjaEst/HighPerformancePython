# T1.1 Implement the map, plot the evolution of x

```python
## Evolution
def bifurcation(x, r):
    return r * x * (1 - x)

@interact(r = 3.3, n=1, m=20)
def evolution(r, n, m):
    res = []
    x = np.random.random()
    for _ in range(n):
        x = bifurcation(x, r)
    for _ in range(m):
        x = bifurcation(x, r)
        res.append(x)
    plt.plot(res)  
```
